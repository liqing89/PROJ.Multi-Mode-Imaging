```python
# 读取目标
data = np.loadtxt(filename, dtype=str)
model_Name = data[:,0]
model_Type = data[:,1]
# 设置运行文件夹
save_folder = '/home/data1/user/PiLiangHuaExperiment/dataset_structure/'
# 开始生产

for i in range(len(model_Name)):
    model_info_folder = '/home/lij/PILIANGHUAEXPERIMENT/35_Targets_POV_and_XML/'+model_Name[i]+'/'
    if model_Name[i] == 'C17' or model_Name[i] == 'B52' or model_Name[i] == 'B2' or model_Name[i] == 'B1B' or model_Name[i] == 'AC130':
        Rho_list = [1, 0.5, 0.3]
    else:
        Rho_list = [0.5,0.3,0.2]
    for polarization in ['HH','VV']:
        for wave_band in ['X','C']:
            for Rho in Rho_list: # 飞机分辨率
                if model_Type[i] == 'JC':
                    if model_Name[i] == 'ZHBJC':
                        distributionx = 455
                        distributiony = 455
                    elif model_Name[i] == 'HF':
                        distributionx = 505
                        distributiony = 505
                    else:
                        distributionx = 275
                        distributiony = 275
                    x_cut = distributionx-5
                    y_cut = distributiony-5
                    RayH = y_cut/Rho
                    RayW = x_cut/Rho
                    beta_range = 3
                    beta_azimuth = 2
                elif model_Type[i] == 'HM':
                    distributionx = 505
                    distributiony = 505
                    if model_Name[i] == 'HMBD':
                        distributionx = 2005
                        distributiony = 2005
                    x_cut = distributionx-5
                    y_cut = distributiony-5
                    RayH = y_cut/Rho
                    RayW = x_cut/Rho
                    beta_range = 3
                    beta_azimuth = 2
                elif model_Type[i] == 'FJ':
                    distributionx = 95
                    distributiony = 95
                    x_cut = distributionx-5
                    y_cut = distributiony-5
                    RayH = y_cut/Rho
                    RayW = x_cut/Rho
                    beta_range = 3
                    beta_azimuth = 2
                elif model_Type[i] == 'TK':
                    distributionx = 35
                    distributiony = 35
                    x_cut = distributionx-5
                    y_cut = distributiony-5
                    RayH = y_cut/Rho
                    RayW = x_cut/Rho
                    beta_range = 4
                    beta_azimuth = 2.5
                else:
                    pass
                current_save_folder = save_folder + model_Type[i] + '/' + model_Name[i] + '/' + polarization + '/' + wave_band + '/' + str(Rho) + '_' + str(Rho) + '/'
                mkdataset(model_Name[i], model_info_folder, current_save_folder, distributionx, distributiony, RayH, RayW,
                        model_Type[i], x_cut, y_cut, scanMode, Rho, beta_range, beta_azimuth, wave_band, polarization, squiAng)
```
